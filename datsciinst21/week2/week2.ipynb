```python
#Exercise 1
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Exercise 2
response = requests.get('https://www.bbc.com/news')
contents = response.text
```

```python
#print(contents)
```

```python
#Exercise 3, 4, 5
soup = BeautifulSoup(contents, 'html.parser')
```

```python
links = soup.find_all('h3', class_="gs-c-promo-heading__title gel-pica-bold nw-o-link-split__text") #headlines section
for link in links:
    print(link.text)
```

```python
#Exercise 6
head = [] 
summ = []
all_stories = soup.find_all('div', class_= 'gs-c-promo-body')
for story in all_stories:
    headline = story.find('h3')
    head.append(headline.text)
    print("Headline:", headline.text)
    #region = story.find('span', class_= 'aria-hidden')
    #print(span.text)
    summary = story.find('p')
    if summary:
        print("Summary:", summary.text)
        summ.append(summary.text)
    else:
        summ.append(" ")
    print("")
```

```python
#Exercise 7    
d= {'Headline' : head, 'Summary' : summ} 
df = pd.DataFrame(data=d)
df.to_csv(r'C:\Users\Jolly\Desktop\export_dataframe.csv', index = False, header=True)
```
