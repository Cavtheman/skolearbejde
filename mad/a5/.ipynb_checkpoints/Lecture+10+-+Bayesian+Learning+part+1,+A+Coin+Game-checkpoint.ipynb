# Lecture 10: Bayesian Learning part 1, A Coin Game

In this lecture we go through the coin game example from Chapter 3.1 in Rogers & Girolami.

```python
# Import packages
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

from scipy.stats import binom
from scipy.stats import beta

import matplotlib
matplotlib.rcParams['figure.figsize'] = (20,10)


matplotlib.rc('font', size=20)
matplotlib.rc('axes', titlesize=20)

from matplotlib import cm                      # Import colormaps
```

The coin game works as follows:
* The customer pays 1 pound
* A coin is tossed 10 times
* If the coin lands heads on 0-6 occasions, the customer gets 2 pounds
* If the coin lands heads on 7-10 occasions, the stall owner keeps their money

The binomial distribution describes the probability of a certain number (y) of heads in N binary events:

$P(Y=y) = $
${N}\choose{y} $
$r^y (1-r)^{N-y}$

Setting $r=0.5$ for a fair coin, and with $N=10$, we obtain the following probability distribution (pmf):

```python
N = 10
r = 0.5

x = np.arange(0,11,1)

plt.bar(x, binom.pmf(x, N, r))
plt.title('Binomial PMF')
plt.xlabel('y')
```

We can compute the probability of doubling your money, i.e. $P(Y \le 6)$:

```python
binom.cdf(6,N,r)
```

We can also compute the expected return from playing the game. Let X be a random variable taking values 1 (winning) or 0 (losing); now

$P(X = 1) = P(Y \le 6)$,

$P(X = 0) = P(Y > 6)$.

Let $f(X)$ be a function giving the return of the game. Thus, if we win, we get a return of 

$f(1) = 2$;

if we lose, we get a return of 

$f(0) = 0$. 

Thus the expected return becomes

$f(1)P(X=1) + f(0)P(X=0) = 2P(Y \le 6) + 0P(Y > 6)$, 

That is, the expected return is:

```python
2*binom.cdf(6,N,r)
```

Since it costs $1$ pound to participate, your expected win is 0.66 pounds.

### Perhaps our assumptions are wrong?
This seems like a bad business model. Perhaps our analysis is based on incorrect assumptions?

Our analysis is based on two assumptions:
* That the number of heads is a random variable following the binomial distribution with $r$ the probability of head per toss
* That the coin is fair, i.e. $r=0.5$.

The binomial distribution is hard to argue, but perhaps the coin is not fair?

We will follow the Bayesian model from Chapters 3.1-3.3 in Rogers & Girolami. That will require:
* An implemented data generator for drawing samples
* An implemented posterior pdf
* Expectation and variance of the posterior

**Before we move on:** We discuss possible priors and derive the formula for the posterior on the board.

** Moving on:** Setting up to experiment with different priors. Remember, we need:
* An implemented data generator for drawing samples
* An implemented posterior pdf
* Expectation and variance of the posterior

A data generating model -- we assume the data is actually generated from a binomial distribution with r=0.9

```python
# Here, 0 = tails, 1 = heads
def draw_sample(r, num):
    return np.random.binomial(1,r,num)
```

Implement an evaluation of the prior pdf for plotting

```python
def prior(x, prior_alpha, prior_beta):
    return beta.pdf(x, prior_alpha, prior_beta)
```

Implement the analytical posterior

```python
def posterior(x, prior_alpha, prior_beta, sample):
    N = len(sample)
    y_N = sum(sample)           # Number of heads
    delta = y_N + prior_alpha
    gamma = N - y_N + prior_beta
    return beta.pdf(x, delta, gamma)
```

Settings for experiments -- note that you can generate your own set of samples to play around. We will run with the one from the book.

```python
# Draw a sample to be used throughout
#n_sample = 20                                        # Self-generated samples
#sample = draw_sample(0.9,n_sample)                   # Self-generated samples
n_sample = 10                                         # Sample from book
sample = [1,0,1,1,1,1,0,0,0,1]                        # Sample from book
#print('Sample:', sample)

# Select sample sizes for which the posterior is to be plotted
#PlottedSampleSizes = [1,2,4,9,19]
PlottedSampleSizes = [1,2,3,4,10]

# Generate x axis samples to be used for plotting prior and posteriors
xsamples = np.linspace(0,1,101)

# Color settings for plotting
colors = colors = cm.cool(np.linspace(0,1,np.max(PlottedSampleSizes)+1))
```

### Case 1: No prior knowledge (uniform prior)

```python
# Set parameters of prior
prior_alpha = 1
prior_beta = 1

# Compute the values of the prior and plot its graph
PriorVals = prior(xsamples, prior_alpha, prior_beta)
plt.plot(xsamples, PriorVals)
plt.title('The uniform prior')
plt.xlabel('r')
plt.ylabel('p(r)')
```

```python
# Compute the posterior including iteratively more points from the sample
PosteriorVals = np.zeros([n_sample, len(xsamples)])
for i in range(n_sample):    
    PosteriorVals[i,:] = posterior(xsamples, prior_alpha, prior_beta, sample[:i+1])
    
# Plot the prior    
plt.plot(xsamples, PriorVals, label='Prior', color='black')

# Plot the posterior with increasing numbers of samples
for i in PlottedSampleSizes:
    plt.plot(xsamples, PosteriorVals[i-1,:], label='Posterior after '+str(i)+' samples', color=colors[i])
    
plt.legend()    
plt.xlabel('r')
plt.ylabel('P(r | sample)')
plt.title('Posterior with a uniform (uninformative) prior')
```

### Case 2: We assume a fair coin

```python
prior_alpha = 50
prior_beta = 50

# Compute the values of the prior for plotting
PriorVals = prior(xsamples, prior_alpha, prior_beta)

plt.plot(xsamples, PriorVals)
plt.title('The "fair coin" prior')
plt.xlabel('r')
plt.ylabel('p(r)')
```

```python
# Compute the posterior including iteratively more points from the sample
PosteriorVals = np.zeros([n_sample, len(xsamples)])
for i in range(n_sample):    
    PosteriorVals[i,:] = posterior(xsamples, prior_alpha, prior_beta, sample[:i+1])
    
# Plot the prior    
plt.plot(xsamples, PriorVals, label='Prior', color='black')

# Plot the posterior with increasing numbers of samples
for i in PlottedSampleSizes:
    plt.plot(xsamples, PosteriorVals[i-1,:], label='Posterior after '+str(i)+' samples', color=colors[i])

plt.legend()        
plt.xlabel('r')
plt.ylabel('P(r | sample)')
plt.title('Posterior with a fair coin prior')
```

### Case 3: We assume a biased coin

```python
prior_alpha = 5
prior_beta = 1

# Compute the values of the prior for plotting
PriorVals = prior(xsamples, prior_alpha, prior_beta)

plt.plot(xsamples, PriorVals)
plt.title('The "biased coin" prior')
plt.xlabel('r')
plt.ylabel('p(r)')
```

```python
# Compute the posterior including iteratively more points from the sample
PosteriorVals = np.zeros([n_sample, len(xsamples)])
for i in range(n_sample):    
    PosteriorVals[i,:] = posterior(xsamples, prior_alpha, prior_beta, sample[:i+1])
    
# Plot the prior    
plt.plot(xsamples, PriorVals, label='Prior', color='black')

# Plot the posterior with increasing numbers of samples
for i in PlottedSampleSizes:
    plt.plot(xsamples, PosteriorVals[i-1,:], label='Posterior after '+str(i)+' samples', color=colors[i])
    
plt.legend()     
plt.xlabel('r')
plt.ylabel('P(r | sample)')
plt.title('Posterior with a biased coin prior')
```
